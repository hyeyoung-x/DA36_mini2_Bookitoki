```python
book_df=pd.read_csv('data/book_review_data.csv')
```

```python
# 1) í•„ìš”í•œ ì»¬ëŸ¼ ë„ì¶œ
# 2) í˜•ë³€í™˜ (float > int)

book_df=book_df[['user_id',"age","rating","book_title",'book_author','year_of_publication','img_m','Summary','Category','country']]
book_df[['age',"year_of_publication"]]=book_df[['age',"year_of_publication"]].astype(int)
```

```python
# ìµœì†Œ í‰ê°€ ê°œìˆ˜ ( í‰ê°€ê°€ 30ê°œë³´ë‹¤ ë§Žì€ ì±…ë§Œ ì¶”ì¶œ )
counts=book_df['book_title'].value_counts()
title = counts[counts > 30].index
book_filter_df = book_df[book_df['book_title'].isin(title)]
book_filter_df.value_counts()
```

```python
# ìµœì†Œ í‰ê°€ ê°œìˆ˜ ( í‰ê°€ê°€ 20ê°œë³´ë‹¤ ë§Žì€ ì‚¬ìš©ìžë§Œ ì¶”ì¶œ )
counts=book_df["user_id"].value_counts()
user_id_=counts[counts>20].index
book_filter_df = book_filter_df[book_filter_df['user_id'].isin(user_id_)]
```

```python
# ë¬´ì˜ë¯¸í•œ 0ì  í‰ê°€ ì œê±°
rating_zero=book_filter_df[book_filter_df['rating']==0].index
book_filter_df.drop(rating_zero, axis=0, inplace=True)
book_filter_df
```

```python
# book_filter_df['Category'].value_counts()
```

```python
# ì¹´í…Œê³ ë¦¬ ì†Œë¬¸ìžë¡œ ë§Œë“¤ê¸° / ì•žë’¤ ê³µë°± ì œê±°/ ì¤‘ë°± ê³µë°± í‘œì¤€í™” / íŠ¹ìˆ˜ë¬¸ìž ì œê±°

# book_filter_df['Category'] == book_filter_df['Category'].str.lower()
# book_filter_df['Category']

def preprocess_category(text):
    return(
        text
        .str.lower()
        .str.strip()  
        .str.replace(r'\s+', ' ', regex=True)
        .str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
    )

book_filter_df['Category']=preprocess_category(book_filter_df['Category'])

book_filter_df['Category'].value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
import pandas as pd

# 1. CountVectorizer ê°ì²´ ìƒì„± ë° ë°ì´í„° ë²¡í„°í™”
count_vectorizer = CountVectorizer()
category_vecs = count_vectorizer.fit_transform(book_filter_df['Category'])
print(category_vecs.shape)
print(count_vectorizer.vocabulary_) # (365778, 1092) (ì¹´í…Œê³ ë¦¬ìˆ˜, ë‹¨ì–´ìˆ˜)
category_vec_vocab = pd.DataFrame(count_vectorizer.vocabulary_,index=['index']).T
category_vec_vocab
```

```python
category_vecs.toarray()[:20]
```

```python
# ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ì¸¡ì •
from sklearn.metrics.pairwise import cosine_similarity

# batchì°¨ì›ì˜ ë°°ì—´ë“¤ì˜ ìœ ì‚¬ë„ ê³„ì‚°
category_sim = cosine_similarity(category_vecs,category_vecs)
```

```python
# í…ìŠ¤íŠ¸ êµ°ì§‘í™” í•˜ê¸°
# ì¹´í…Œê³ ë¦¬ 380ê°œ -> 20ê°œë¡œ êµ°ì§‘í™”
import re
from nltk import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

unique_category = book_filter_df['Category'].unique()
unique_category = preprocess_category(pd.Series(unique_category)).tolist()

# ë²¡í„°í™”
tfidf_vectorizer = TfidfVectorizer(min_df=2, max_df=0.5)
tfidf_matrix = tfidf_vectorizer.fit_transform(unique_category)

# kmeans
kmeans = KMeans(n_clusters=30, random_state=0)
kmeans.fit(tfidf_matrix)
clusters = kmeans.labels_
temp_df=pd.DataFrame({
    'cluster':clusters,
    'category':unique_category,
})

representative_genres =dict(temp_df.groupby('cluster')['category'].first())

reversed_representative_genres = {value: key for key, value in representative_genres.items()}

book_filter_df['category_clustered'] = book_filter_df['Category'].map(reversed_representative_genres)
book_filter_df
print(book_filter_df['category_clustered'].isnull().sum())

unmapped_categories = book_filter_df[book_filter_df['category_clustered'].isnull()]['Category'].unique()
print("ë§¤í•‘ë˜ì§€ ì•Šì€ Category ê°’ë“¤:", unmapped_categories)

```
